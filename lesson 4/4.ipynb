### Задача 1

#### О непрерывной равномерно распределенной случайной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Формула вычисления дисперсии для равномерного распределения случайной величины: D(X) = (b-a) ** 2 / 12 => (b - a) ** 2 = 12 * D(X) => b = sqrt(12 * D(X)) + a

```python
D = 0.2
a = 0.5
print (f'правая граница b равна {np.sqrt(12 * D) + a}')
```

#### 4. от 166 см до 182 см?

```python
#Используя теорию трех сигм можно определить вероятности (я присылал в телегу фото своих каракулей со схемой).
sigma_1 = 0.68
sigma_2 = 0.954
sigma_3 = 0.9972
```

```python
print(f'P1 = {(1 - sigma_1)/2}')
print(f'P2 = {(1 - sigma_2)/2}')
print(f'P3 = {sigma_2 - (sigma_2 - sigma_1)/2}')
print(f'P4 = {sigma_1}')
print(f'P5 = {sigma_2}')
print(f'P6 = {(sigma_3 - sigma_2)/2 - 0}')
print(f'P7 = {1 - sigma_3}')
print(f'P8 = {(sigma_3 - sigma_1)/2}')
```

```python
print (f'P1 = {1 - st.norm.cdf(182, loc=174, scale=8)}')
print (f'P2 = {1 - (st.norm.cdf(190, loc=174, scale=8))}')
print (f'P3 = {(st.norm.cdf(190, loc=174, scale=8)) - (st.norm.cdf(166, loc=174, scale=8))}')
print (f'P4 = {(st.norm.cdf(182, loc=174, scale=8)) - (st.norm.cdf(166, loc=174, scale=8))}')
print (f'P5 = {(st.norm.cdf(190, loc=174, scale=8)) - (st.norm.cdf(158, loc=174, scale=8))}')
print (f'P6 = {(st.norm.cdf(150, loc=174, scale=8)) + (1 - (st.norm.cdf(190, loc=174, scale=8)))}')
print (f'P7 = {(st.norm.cdf(150, loc=174, scale=8)) + (1 - (st.norm.cdf(198, loc=174, scale=8)))}')
print (f'P8 = {(st.norm.cdf(166, loc=174, scale=8))}')
```
